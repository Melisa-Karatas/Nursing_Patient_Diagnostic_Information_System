```python
import sqlite3

class Hasta:
    def __init__(self, servis, ad, soyad, tc, yas, tani, alerji, ameliyat, itaki_puani, glaskow):
        self.servis = servis
        self.ad = ad
        self.soyad = soyad
        self.tc = tc
        self.yas = yas
        self.tani = tani
        self.alerji = alerji
        self.ameliyat = ameliyat
        self.itaki_puani = itaki_puani
        self.glaskow = glaskow
        
    def bilgi_goster(self):
            print("GENEL BİLGİLER")
            print(f"""
                Hastanın Yattığı Servis: {self.servis}
                Hastanın Adı - Soyadı: {self.ad} {self.soyad}
                Hastanin Yasi: {self.yas}
                Hastanın Tc Kimlik Numarası: {self.tc}
                Tanı: {self.tani}
                Hastanın Varsa Alerjileri: {self.alerji}
                Varsa Yapılan Ameliyat: {self.ameliyat}
                İtaki Puanı: {self.itaki_puani}
                Glaskow Puanı: {self.glaskow}
            """)
        
    def bilgi_guncelle(self):
        servis_guncelle = input("Lütfen yeni servis bilgisini giriniz. ")
        ad_guncelle = input("Lütfen yeni ad bilgisini giriniz. ")
        soyad_guncelle = input("Lütfen yeni soyad bilgisini giriniz. ")
        tc_guncelle = input("Lütfen yeni tc kimlik bilgisini giriniz. ")
        yas_guncelle = input("Lütfen yeni yaş bilgisini giriniz. ")
        tani_guncelle = input("Lütfen yeni tanı bilgisini giriniz. ")
        alerji_guncelle = input("Lütfen yeni alerji bilgisini giriniz. ")
        ameliyat_guncelle = input("Lütfen yeni ameliyat bilgisini giriniz. ")
        itaki_guncelle = input("Lütfen yeni itaki puan bilgisini giriniz. ")
        glaskow_guncelle = input("Lütfen yeni glaskow puan bilgisini giriniz. ")
        
        if servis_guncelle:
            self.servis = servis_guncelle
        if ad_guncelle:
            self.ad = ad_guncelle
        if soyad_guncelle:
            self.soyad = soyad_guncelle
        if tc_guncelle:
            self.tc = tc_guncelle
        if yas_guncelle:
            self.yas = yas_guncelle
        if tani_guncelle:
            self.tani = tani_guncelle
        if alerji_guncelle:
            self.alerji = alerji_guncelle
        if ameliyat_guncelle:
            self.ameliyat = ameliyat_guncelle
        if itaki_guncelle:
            self.itaki_puani = itaki_guncelle
        if glaskow_guncelle:
            glaskow = glaskow_guncelle
        print("Hasta bilgileri güncellenmiştir.")
        
class Anta:
    def __init__(self, gun, ay, yil, saat, tc, ates, nabiz, sis_kanb, diya_kanb, agri):
        self.gun = gun
        self.ay = ay
        self.yil = yil
        self.saat = saat
        self.tc = tc
        self.ates = ates
        self.nabiz = nabiz
        self.sis_kanb = sis_kanb
        self.diya_kanb = diya_kanb
        self.agri = agri
    
    def bilgi_goster(self):
        print("ANTA BİLGİLERİ")
        print(f"""
            Tarih: {self.yil}/{self.ay}/{self.saat}  Saat:{self.tc}
            Hastanın Tc Kimlik Numarası: {self.gun}
            Ateş: {self.ates}, Nabız: {self.nabiz}, Kan Basıncı: {self.sis_kanb}/{self.diya_kanb}, 
            Ağrı(var ise 10 üzerinden puan): {self.agri}
        """)
        
    def bilgi_guncelle(self):
        gun_guncelle = input("Lütfen yeni gün bilgisini giriniz. ")
        ay_guncelle = input("Lütfen yeni ay bilgisini giriniz. ")
        saat_guncelle = input("Lütfen yeni saat bilgisini giriniz. ")
        yeni_tc = input("Lütfen yeni tc bilgisini giriniz. ")
        ates_guncelle = input("Lütfen yeni ateş bilgisini giriniz. ")
        nabiz_guncelle = input("Lütfen yeni nabız bilgisini giriniz. ")
        sis_kanb_guncelle = input("Lütfen yeni sistolik kan basıncı bilgisini giriniz. ")
        diya_kanb_guncelle = input("Lütfen yeni diyastolik kan basıncı bilgisini giriniz. ")
        agri_guncelle = input("Lütfen yeni ağrı bilgisini giriniz. ")
        
        if gun_guncelle:
            self.gun = gun_guncelle
        if ay_guncelle:
            self.ay = ay_guncelle
        if saat_guncelle:
            self.saat = saat_guncelle
        if yeni_tc:
            self.tc = yeni_tc
        if ates_guncelle:
            self.ates = ates_guncelle
        if nabiz_guncelle:
            self.nabiz = nabiz_guncelle
        if sis_kanb_guncelle:
            self.sis_kanb = sis_kanb_guncelle
        if diya_kanb_guncelle:
            self.diya_kanb = diya_kanb_guncelle
        if agri_guncelle:
            self.agri = agri_guncelle
        print("Hasta bilgileri güncellenmiştir.")
        
class Ilac:
    def __init__(self, tc, guncel_ilaclar, duzenli_ilaclar):
        self.tc = tc
        #self.ilac_sayisi
        self.guncel_ilaclar = guncel_ilaclar
        self.duzenli_ilaclar = duzenli_ilaclar
        
    def bilgi_goster(self):
        print("HASTANIN KULLANDIĞI İLAÇLAR") 
        print(f"""
            Hastanın güncel olarak kullandığı ilaçlar: {self.guncel_ilaclar}
            Hastanın düzenli olarak kullandığı ilaçlar: {self.duzenli_ilaclar}
        """)
        
    def bilgi_guncelle(self):
        guncel_ilac_duzenle = input("Lütfen yeni güncel ilaç bilgisini giriniz. ")
        duzenli_ilac_duzenle = input("Lütfen yeni düzenli ilaç bilgisini giriniz. ")
        
        if guncel_ilac_duzenle:
            self.guncel_ilaclar = guncel_ilac_duzenle
        if duzenli_ilac_duzenle:
            self.duzenli_ilaclar = duzenli_ilac_duzenle
        print("Hastaya ait ilaç bilgileri güncellenmiştir.")
        
        
class HastaSistemTanilama:
    def __init__(self):
        self.connect = sqlite3.connect('hemsirelik_sistem_tanilama.db')
        self.cursor = self.connect.cursor()
        self.create_tables()
        
    def create_tables(self):
        self.cursor.execute('''
            CREATE TABLE IF NOT EXISTS Hasta (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                servis VARCHAR(300),
                ad VARCHAR(100) ,
                soyad VARCHAR(100),
                tc_numarasi VARCHAR(11),
                yas INTEGER,
                tani VARCHAR(500),
                alerji VARCHAR(500),
                ameliyat VARCHAR(500),
                itaki_puani INTEGER ,
                glaskow INTEGER 
            )
        ''')

        self.cursor.execute('''
            CREATE TABLE IF NOT EXISTS Anta (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                tc VARCHAR(11) NOT NULL,
                gun INTEGER NOT NULL,
                ay INTEGER NOT NULL,
                yil INTEGER NOT NULL,
                saat INTEGER NOT NULL,
                ates FLOAT NOT NULL,
                nabiz INTEGER NOT NULL,
                sis_kanb INTEGER NOT NULL,
                diya_kanb INTEGER NOT NULL,
                agri INTEGER NOT NULL,
                FOREIGN KEY (tc) REFERENCES Hasta(tc_numarasi)
            )
        ''')
        
        self.cursor.execute('''
            CREATE TABLE IF NOT EXISTS Ilac (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                tc_kimlik VARCHAR(11) NOT NULL,
                guncel_ilaclar VARCHAR(600),
                duzenli_ilaclar VARCHAR(600),
                FOREIGN KEY (tc_kimlik) REFERENCES Hasta(tc_numarasi)
             )   
        ''')
        
        self.connect.commit()
        
    def hasta_ekle(self, hasta):
        self.cursor.execute('''
            INSERT INTO HASTA (servis, ad, soyad, tc_numarasi, yas, tani, alerji, ameliyat, itaki_puani, glaskow)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        ''', (hasta.servis, hasta.ad, hasta.soyad, hasta.tc, hasta.yas, hasta.tani, hasta.alerji, 
              hasta.ameliyat, hasta.itaki_puani, hasta.glaskow))
        self.connect.commit()
        print(f" {hasta.ad} {hasta.soyad} adlı hasta yatışı gerçekleşti.")
        
    def anta_ekle(self, anta):
        self.cursor.execute('''
            INSERT INTO ANTA (gun, ay, yil, saat, tc, ates, nabiz, sis_kanb, diya_kanb, agri)
            VALUES(?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        ''', (anta.gun, anta.ay, anta.yil, anta.saat, anta.tc, anta.ates, anta.nabiz, anta.sis_kanb, anta.diya_kanb, anta.agri))
        self.connect.commit()
        print(f" {hasta.ad} {hasta.soyad} adlı hasta ait anta verileri kaydedildi.")
        
    def ilac_ekle(self, ilac):
        self.cursor.execute('''
            INSERT INTO ILAC (tc_kimlik, guncel_ilaclar, duzenli_ilaclar) VALUES(?, ?, ?)
        ''', (ilac.tc, ilac.guncel_ilaclar, ilac.duzenli_ilaclar))
        self.connect.commit()
        print(f"{hasta.ad} {hasta.soyad} adlı hastaya ait ilaç verileri kayıt edildi.")
        
    def Glaskow_skalasi(self, tc):
        print("Lütfen aşağıya hastanın durumuna göre uygun olan puanları giriniz.")
        goz = int(input("""
        Hasta'nın gözleri:
        *Spontan açık ve normal hareket var ise 4 puan
        *Sözel uyarana açık ise 3 puan
        *Ağrılı uyarana açık ise 2 puan
        *Yanıt yok ise 1 puan giriniz.
        """))

        sozel_yanit = int(input("""
        Hasta:
        *Oryante/Koopere ise 5 puan
        *Konfüze ise 4 puan
        *Uygun olmayan kelimeler kullanıyor ve anlamsız cevaplar veriyor ise 3 puan
        *Tanımlanamayan kelime ve ses çıkarıyorsa 2 puan
        *Ses yok ise 1 puan giriniz.
        """))

        motor_yanit = int(input("""
        Hasta:
        *Emirlere uyuyor ve ekstremiteleri hareket ettiriyorsa 6 puan
        *Ağrılı uyaran verilen ekstremiteyi hareket ettiriyorsa 5 puan
        *Ağrılı uyarandan ekstremiteyi çekiyorsa 4 puan
        *Dekortike postür ve anormal fleksiyonda ise 3 puan
        *Deserebre postürde ve ekstansör yanıt veriyorsa 2 puan
        *Yanıt yok ise 1 puan giriniz.
        """))

        glaskow_puani = goz + sozel_yanit + motor_yanit
        if 3<= glaskow_puani <=7:
            print("Hastada derin nörolojik hasar vardır.")
        elif 8<= glaskow_puani <= 11:
            print("Hastada orta düzey nörolojik hasar vardır.")
        elif 12<= glaskow_puani <= 15:
            print("Hastada hafif düzeyde nörolojik hasar vardır.")
            
    def hasta_goster(self):
        rows = self.cursor.execute('SELECT * FROM HASTA').fetchall()
        for row in rows:
            hasta = Hasta(row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10])
            hasta.bilgi_goster()
            
    def anta_goster(self):
        rows = self.cursor.execute('SELECT * FROM ANTA').fetchall()
        for row in rows:
            anta = Anta(row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10])
            anta.bilgi_goster()
            
    def ilac_goster(self):
        rows = self.cursor.execute('SELECT * FROM ILAC').fetchall()
        for row in rows:
            ilac = Ilac(row[1], row[2], row[3])
            ilac.bilgi_goster()
            
    def hasta_sil(self, tc):
        self.cursor.execute('DELETE FROM HASTA WHERE tc_numarasi = ?', (tc,))
        self.connect.commit()
        print(f"{ad} {soyad} adlı hasta kaydı başarılı bir şekilde silindi.")
        
    def anta_sil(self, tc):
        self.cursor.execute('DELETE FROM ANTA WHERE tc = ?', (tc,))
        self.connect.commit()
        print(f"{ad} {soyad} adlı hastaya ait anta verileri başarılı bir şekilde silindi.")
        
    def ilac_sil(self, tc):
        self.cursor.execute('DELETE FROM ILAC WHERE tc_kimlik = ?', (tc,))
        self.connect.commit()
        print(f"{tc} tc numaralı hastaya ait ilaç verileri başarılı bir şekilde silindi.")
        
    def hasta_ara(self, tc):
        row = self.cursor.execute('SELECT * FROM Hasta WHERE tc_numarasi = ?',(tc,)).fetchone()
        if row:
            return Hasta(row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10])
        return None
        
    def anta_ara(self, tc):
        row = self.cursor.execute('SELECT * FROM Anta WHERE tc = ?',(tc,)).fetchone()
        if row:
            return Anta(row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10])
        return None
    
    def ilac_ara(self, tc):
        row = self.cursor.execute('SELECT * FROM ILAC WHERE tc_kimlik=?',(tc,)).fetchone()
        if row:
            return Ilac(row[1], row[2], row[3])
        return None
    
    def hasta_guncelleme(self, tc):
        row = self.cursor.execute('SELECT * FROM Hasta WHERE tc_numarasi = ?', (tc,)).fetchone()
        if row:
            hasta = Hasta(row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10])
            hasta.bilgi_guncelle()
            self.cursor.execute('''
                UPDATE Hasta
                SET servis = ?, ad = ?, soyad = ?, tc_numarasi = ?, yas = ?, tani = ?, 
                alerji = ?, ameliyat = ?, itaki_puani = ?, glaskow = ?
                WHERE tc_numarasi = ?
            ''', (hasta.servis, hasta.ad, hasta.soyad, hasta.tc, hasta.yas, hasta.tani, hasta.alerji, 
              hasta.ameliyat, hasta.itaki_puani, hasta.glaskow, tc))
            self.connect.commit()
        else:
            print("Hasta bulunamadı.")
            
    def anta_guncelleme(self, tc):
        row = self.cursor.execute('SELECT * FROM Anta WHERE tc = ?',(tc,)).fetchone()
        if row:
            anta = Anta(row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10])
            anta.bilgi_guncelle()
            self.cursor.execute('''
                Update Anta
                SET gun = ?, ay = ?, yil = ?, saat = ?, tc = ?, ates = ?, nabiz = ?, sis_kanb = ?, diya_kanb = ?, agri = ?
                WHERE tc = ?
            ''', (anta.gun, anta.ay, anta.yil, anta.saat, anta.tc, anta.ates, anta.nabiz, 
                  anta.sis_kanb, anta.diya_kanb, anta.agri, tc))
            self.connect.commit()
        else:
            print("Aradığınız tc kimlik numarasına ait ANTA verileri bulunamadı.")
            
    def ilac_guncelleme(self, tc):
        row = self.cursor.execute('SELECT * FROM Ilac WHERE tc_kimlik = ?', (tc,)).fetchone()
        if row:
            ilac = Ilac(row[1], row[2], row[3])
            ilac.bilgi_guncelle()
            self.cursor.execute('''
                Update Ilac
                SET tc_kimlik = ?, guncel_ilaclar=?, duzenli_ilaclar= ? WHERE tc_kimlik = ?
            ''',(ilac.tc, ilac.guncel_ilaclar, ilac.duzenli_ilaclar, tc))
            self.connect.commit()
        else:
            print("Aradığınız tc kimlik numarasına ait ilaç verileri bulunamadı.")
            
            
    def close_connection(self):
        self.connect.close()
        
hasta_sistem_tanilama = HastaSistemTanilama()

while True:
    print(20 * "*", "HEMŞİRELİK HASTA VERİ KAYIT SİSTEMİ", 20 * "*")
    print("""
        Lütfen yapacağınız işlemin numarasını giriniz:
        
        1. YENİ HASTA YATIŞI
        2. ANTA GİRİŞİ
        3. İLAÇ GİRİŞİ
        4. HASTA LİSTESİ
        5. ANTA LİSTESİ
        6. İLAÇ LİSTESİ 
        7. HASTA ÇIKIŞI/ VERİ SİL
        8. ANTA VERİSİ SİL
        9. İLAÇ SİL
       10. HASTA GÜNCELLE
       11. ANTA GÜNCELLE
       12. İLAÇ GÜNCELLE
       13. HASTA ARA
       14. ANTA ARA 
       15. İLAÇ ARA
       16. SİSTEMDEN ÇIK (Q/q)
    """)
    islem = input("Lütfen yapacağınız işlemin numarasını giriniz: ")

    if islem == "1":
        servis = input("Hastanın Yattığı Servis: ")
        ad = input("Ad: ")
        soyad = input("Soyad: ")
        tc_numarasi = input("Tc: ")
        yas = input("Yaş: ")
        tani = input("Tanı: ")
        alerji = input("Hastanın Varsa Alerjileri: ")
        ameliyat = input("Varsa Yapılan Ameliyat: ")
        itaki_puani = input("İtaki Puanı: ")
        glaskow = input("Glaskow Puanı: ")
        #glaskow = hasta_sistem_tanilama.Glaskow_skalasi(tc)
        hasta = Hasta(servis, ad, soyad, tc_numarasi, yas, tani, alerji, ameliyat, itaki_puani, glaskow)
        hasta_sistem_tanilama.hasta_ekle(hasta)
        
    elif islem == "2":
        gun = input("Gün: ")
        ay = input("Ay: ")
        yil = input("Yıl: ")
        saat = input("Saat: ")
        tc = input("Hastanın Tc Kimlik Numarası: ")
        ates = input("Ateş: ")
        nabiz = input("Nabız: ")
        sis_kanb = input("Sistolik Kan Basıncı: ")
        diya_kanb = input("Diyastolik Kan Basıncı: ")
        agri = input("Ağrı(var ise 10 üzerinden puan): ")
        
        hasta= hasta_sistem_tanilama.hasta_ara(tc)
        if hasta:
            anta = Anta(gun, ay, yil, saat, tc, ates, nabiz, sis_kanb, diya_kanb, agri)
            hasta_sistem_tanilama.anta_ekle(anta)
        else:
            print("Girmiş olduğunuz bilgilere ait hasta bulunamadı. Lütfen hasta ekledikten sonra anta verisi giriniz.")
            
    elif islem == "3":
        tc = input("Hastanın Tc Kimlik Numarası: ")
        guncel_ilaclar = input("Hastanın güncel olarak kullandığı ilaçların adını yazınız: ")
        duzenli_ilaclar = input("Hastanın düzenli olarak kullandığı ilaçların adını yazınız: ")
        
        hasta= hasta_sistem_tanilama.hasta_ara(tc)
        if hasta:
            ilac = Ilac(tc, guncel_ilaclar, duzenli_ilaclar)
            hasta_sistem_tanilama.ilac_ekle(ilac)
        else:
            print("Girmiş olduğunuz bilgilere ait hasta bulunamadı. Lütfen hasta ekledikten sonra ilaç verisi giriniz.")
        
    elif islem == "4":
        hasta_sistem_tanilama.hasta_goster()
        
    elif islem == "5":
        hasta_sistem_tanilama.anta_goster()
        
    elif islem == "6":
        hasta_sistem_tanilama.ilac_goster()
    
    elif islem == "7":
        tc = input("Hastayı sistemden silmek için lütfen tc'sini giriniz: ")
        hasta_sistem_tanilama.hasta_sil(tc)
        
    elif islem == "8":
        tc = input("Anta verilerini sistemden silmek için lütfen hastanın tc'sini giriniz: ")
        hasta_sistem_tanilama.anta_sil(tc)
    
    elif islem == "9":
        tc = input("İlaç verilerini sistemden silmek için lütfen hastanın tc'sini giriniz: ")
        hasta_sistem_tanilama.ilac_sil(tc)
        
    elif islem == "10":
        tc = input("Verilerini güncellemek istediğiniz hastaya ait tc kimlik numarasını giriniz. ")
        hasta_sistem_tanilama.hasta_guncelleme(tc)
        
    elif islem == "11":
        tc = input("Anta verilerini güncellemek istediğiniz hastaya ait tc kimlik numarasını giriniz. ")
        hasta_sistem_tanilama.anta_guncelleme(tc)
        
    elif islem == "12":
        tc = input("İlaç verilerini güncellemek istediğiniz hastaya ait tc kimlik numarasını giriniz. ")
        hasta_sistem_tanilama.ilac_guncelleme(tc)
        
    elif islem == "13":
        tc = input("Verilerini görmek istediğiniz hastaya ait tc kimlik numarasını giriniz: ")
        hasta = hasta_sistem_tanilama.hasta_ara(tc)
        if hasta:
            hasta.bilgi_goster()
        else:
            print("Girmiş olduğunuz tc kimlikli hasta bulunamadı.")
            
    elif islem == "14":
        tc = input("Anta verilerini görmek istediğiniz hastaya ait tc kimlik numarasını giriniz: ")
        anta = hasta_sistem_tanilama.anta_ara(tc)
        if anta:
            anta.bilgi_goster()
        else:
            print("Girmiş olduğunuz tc kimlikli hastaya ait anta verisi bulunamadı.")
    
    elif islem == "15":
        tc = input("İlaç verilerini görmek istediğiniz hastaya ait tc kimlik numarasını giriniz. ")
        ilac = hasta_sistem_tanilama.ilac_ara(tc)
        if ilac:
            ilac.bilgi_goster()
        else:
            print("Girmiş olduğunuz tc kimlikli hastaya ait ilaç verisi bulunamadı.")
        
    elif islem == "16" or islem.lower() == "q":
        print("Sistemden çıkış yaptınız.")
        break
    else:
        print("Lütfen geçerli bir işlem numarası giriniz.")
```
